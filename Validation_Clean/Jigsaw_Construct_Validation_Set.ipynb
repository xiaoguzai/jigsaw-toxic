```python
import pandas as pd
import numpy as np
```

```python
validation = pd.read_csv("/content/drive/MyDrive/Jigsaw/validation_data.csv")
validation["comb"] = validation[["less_toxic", "more_toxic"]].values.tolist()
print(validation[["less_toxic", "more_toxic"]].sort_values("less_toxic"))
print(validation[["less_toxic", "more_toxic"]].sort_values("more_toxic"))
# sort pair into first and second
validation["comb_sorted"] = validation["comb"].apply(lambda x: sorted(x))
# if first less toxic than second, score = 0, else 1
validation["score"] = np.where(validation["comb_sorted"] == validation["comb"], 0, 1)

validation = validation[["worker", "comb_sorted", "score"]]
# Split comb column into first and second columns
validation[['first','second']] = pd.DataFrame(validation["comb_sorted"].tolist(), index= validation.index)
validation = validation.groupby(['first','second']).agg({"score":"mean"}).reset_index()
# Average Score larger than 0.5 is first more toxic than second
validation["more_toxic"] = np.where(validation["score"] >= 0.5, 1, 0)
# Scores are between 0.5 - 1, the higher the score, the more certain the pair is
validation["score"] = np.where(validation["score"]<0.5, 1-validation["score"], validation["score"]).round(4)

validation.sort_values("first")
```

```python
validation.to_csv("/content/drive/MyDrive/Jigsaw/validation_cleaned.csv")
```
