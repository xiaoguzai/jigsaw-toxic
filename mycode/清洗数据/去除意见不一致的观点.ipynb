```python
import pandas as pd
import numpy as np

validation = pd.read_csv("/home/xiaoguzai/程序/kaggle有毒评论对比比赛/validation_data.csv")
```

```python

validation["comb"] = validation[["less_toxic", "more_toxic"]].values.tolist()
print(validation[["less_toxic", "more_toxic"]].sort_values("less_toxic"))
print(validation[["less_toxic", "more_toxic"]].sort_values("more_toxic"))
#validation["comb"]为一个list，[validation["less_toxic"][index],validation["more_toxic"][index]]
validation["comb_sorted"] = validation["comb"].apply(lambda x: sorted(x))
#排序validation["less_toxic"],validation["more_toxic"]
validation["score"] = np.where(validation["comb_sorted"] == validation["comb"], 0, 1)
#less_toxic在前为0,more_toxic在前为1
validation = validation[["worker", "comb_sorted", "score"]]
#去除"less_toxic","more_toxic"
validation[['first','second']] = pd.DataFrame(validation["comb_sorted"].tolist(), index= validation.index)
#validation['first'] = validation["comb_sorted"][0],validation['second'] = validation["comb_sorted"][1]
validation = validation.groupby(['first','second']).agg({"score":"mean"}).reset_index()
#!!!agg:聚合函数，求first,second组成的组，当前的'first','second'占平均内容的多少{"score":"mean"}
validation["more_toxic"] = np.where(validation["score"] >= 0.5, 1, 0)
# Scores are between 0.5 - 1, the higher the score, the more certain the pair is
validation["score"] = np.where(validation["score"]<0.5, 1-validation["score"], validation["score"]).round(4)

validation.sort_values("first")

```

```python
validation = validation[abs(validation['score'] - 1) < 1e-10]
validation.head()
#去除多人意见不一致的部分
```

```python
validation["less_toxic"] = np.where(validation["more_toxic"] == 0,validation['first'],validation['second'])
validation["more"] = np.where(validation["more_toxic"] == 1,validation['first'],validation['second'])
validation["more_toxic"] = validation["more"]
#将validation中的more_toxic和less_toxic的内容
```
